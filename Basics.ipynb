```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head()
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no = {"Yes": True, 
          "No": False
}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
import numpy as np

movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_map)
    
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "rating_2",
        "Unnamed: 11": "rating_3",
        "Unnamed: 12": "rating_4",
        "Unnamed: 13": "rating_5",
        "Unnamed: 14": "rating_6"
        })
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

objects=('I', 'II', 'III', 'IV', 'V', 'VI')
y_pos = np.arange(len(objects))
plt.bar(y_pos, star_wars[star_wars.columns[9:15]].mean(), align='center', alpha=0.75)
plt.title('Star Wars Movie Survey Ratings')
plt.xlabel('Movie Number')
plt.xticks(y_pos, objects)
plt.ylabel('Rating')
plt.show()
```

## Star Wars Ratings

- The first trilogy of movies; IV, V, and VI have better ratings than the second trilogy. Note that a lower rating is better, due to how the question was formed.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(y_pos, star_wars[star_wars.columns[3:9]].sum(), align='center', alpha=0.75)
plt.title('Star Wars Movie Number of Viewers')
plt.xlabel('Movie Number')
plt.xticks(y_pos, objects)
plt.ylabel('Viewer Number')
plt.show()
```

## Star Wars Viewer Numbers

- Episodes V and VI had the highest viewer numbers, which reinforces our results found from analyzing the movie ratings.
- Interestingly, Episode I had more viewers than Episode IV.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```
